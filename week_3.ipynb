```python
class MaxHeap():
    def __init__(self):
        self.values = []

    def __bubbleUp(self):
        idx = len(self.values)-1
        element = self.values[idx]
        while idx > 0 :
            parentIdx = (idx-1)//2
            parent = self.values[parentIdx]
            if element <= parent:
                break
            self.values[parentIdx] = element
            self.values[idx] = parent
            idx = parentIdx

    def __sinkDown(self):
        idx = 0
        length = len(self.values)
        element = self.values[0]

        while True:
            leftChildIdx = 2 * idx + 1
            rightChildIdx = 2 * idx + 2
            swap = None

            if leftChildIdx < length:
                leftChild = self.values[leftChildIdx]
                if leftChild > element:
                    swap = leftChildIdx

            if rightChildIdx < length:
                rightChild= self.values[rightChildIdx]
                if not swap and rightChild > element or swap and rightChild > leftChild:
                    swap = rightChildIdx

            if not swap:
                break
            self.values[idx] = self.values[swap]
            self.values[swap] = element
            idx = swap

    def insert(self, element):
        self.values.append(element)
        self.__bubbleUp()

    def maxValue(self):
        max = self.values[0]
        end = self.values.pop()

        if self.values:
            self.values[0] = end
            self.__sinkDown()
        return max
```

```python
maxHeap = MaxHeap()

for i in range(1,7):
    maxHeap.insert(i)

maxHeap.values
```

```python
maxHeap.maxValue()
```

```python
class MinHeap():
    def __init__(self):
        self.values = []

    def __bubbleUp(self):
        idx = len(self.values)-1
        element = self.values[idx]
        while idx > 0 :
            parentIdx = (idx-1)//2
            parent = self.values[parentIdx]
            if element >= parent:
                break
            self.values[parentIdx] = element
            self.values[idx] = parent
            idx = parentIdx

    def __sinkDown(self):
            idx = 0
            length = len(self.values)
            element = self.values[0]

            while True:
                leftChildIdx = 2 * idx + 1
                rightChildIdx = 2 * idx + 2
                swap = None

                if leftChildIdx < length:
                    leftChild = self.values[leftChildIdx]
                    if leftChild < element:
                        swap = leftChildIdx

                if rightChildIdx < length:
                    rightChild= self.values[rightChildIdx]
                    if not swap and rightChild < element or swap and rightChild < leftChild:
                        swap = rightChildIdx

                if not swap:
                    break
                self.values[idx] = self.values[swap]
                self.values[swap] = element
                idx = swap

    def insert(self, element):
        self.values.append(element)
        self.__bubbleUp()

    def minValue(self):
        min = self.values[0]
        end = self.values.pop()

        if self.values:
            self.values[0] = end
            self.__sinkDown()
        return min
```

```python
minHeap = MinHeap()

for i in range(6,0,-1):
    minHeap.insert(i)

minHeap.values
```

```python
def isMaxHeap(heap):
    length = len(heap.values)
    
    for i in range(0,length):
        parent = heap.values[i]

        leftChildIdx = 2 * i + 1
        rightChildIdx = 2 * i + 2

        if leftChildIdx < length:
            leftChild = heap.values[leftChildIdx]
            if leftChild >= parent:
                return False
        

        if rightChildIdx < length:
            rightChild= heap.values[rightChildIdx]
            if rightChild >= parent:
                return False
                
    return True
               
```
